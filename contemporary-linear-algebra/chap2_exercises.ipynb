```python
import numpy as np
import matplotlib.pyplot as plt

# create an array of x values
x = np.linspace(-10, 10, 400)


# calculate corresponding y values for each line
y1 = (1 - x) / 2
y2 = 2 - 2*x
y3 = (3 - 3*x) / -3

# plot the lines
plt.plot(x, y1, label="x + 2y = 1")
plt.plot(x, y2, label="2x + y = 2")
plt.plot(x, y3, label="3x - 3y = 3")

# Add title and labels
plt.title('Graphs of the equations')
plt.xlabel('x')
plt.ylabel('y')

# display a legend
plt.legend()

# show the plot
plt.show()
```
